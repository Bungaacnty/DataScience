### 4.1.1 Feature Engineering

Goals: akan meng-groupkan data pernegara, lalu kita average-kan.

```python
DatasetFinal
```

```python
target_columns_cluster = [
    'country',
    'year',
    'gdp',
    'population',
    'oil_production',
    'Individuals using the Internet (% of population) - IT.NET.USER.ZS',
    'Life expectancy at birth, total (years) - SP.DYN.LE00.IN',
    'Adjusted net national income per capita (annual % growth) - NY.ADJ.NNTY.PC.KD.ZG',
    'GDP per capita (current US$) - NY.GDP.PCAP.CD'
]

DatasetCluster = DatasetFinal[target_columns_cluster]
DatasetCluster
```

```python
DatasetCluster.drop(
  columns = ['year'],
  inplace = True
)
```

```python
# Mencari rata-rata dari data yang dikelompokkan berdasarkan negara

DatasetClusterFinal = DatasetCluster.groupby('country').mean()
DatasetClusterFinal
```

```python
from pycaret.clustering import *

ModelCluster = setup(
    DatasetClusterFinal,
    normalize=True          # Dia akan mengubah value dalam scale data lebih kecil
)
```

```python
# OOP API
ModelKmeans = ModelCluster.create_model('kmeans')
```

```python
# OOP API
ModelCluster.evaluate_model(ModelKmeans)
```

```python
# OOP API
result = ModelCluster.assign_model(ModelKmeans)
result.head()
```

```python
DataResult = ModelCluster.assign_model(ModelKmeans)
DataResult
```

```python
indonesia_cluster = result[result.index == 'Indonesia']['Cluster'].iloc[0]
print(f"Indonesia belongs to cluster: {indonesia_cluster}")
```

```python
# Create a scatter plot with hue for clusters
sns.scatterplot(
    data=DataResult,
    x='oil_production',
    y='gdp',
    hue='Cluster',
    palette='Set2'
)
```

**Insight**: Indonesia sebagai salah satu negara yang menduduki posisi clustering 2 memiliki makna bahwa Indonesia terkonsentrasi di GDP dan oil_production kecil-menengah.

# 4.2 Supervised Learning

### 4.2.1 Regression

Objective: untuk memprediksi GDP dan Biofuel_consumption

```python
target_regression_1 = ['biofuel_consumption']
input_regression_1 = [
    'Adjusted net national income per capita (annual % growth) - NY.ADJ.NNTY.PC.KD.ZG',
    'Gross savings (% of GDP) - NY.GNS.ICTR.ZS',
    'electricity_generation',
    'Trade (% of GDP) - NE.TRD.GNFS.ZS',
    'Broad money (% of GDP) - FM.LBL.BMNY.GD.ZS',
    'oil_production'
]
all_column_regression = input_regression_1 + target_regression_1  # Untuk mengkombine all columns
all_column_regression
```

```python
from pycaret.regression import *
DatasetRegression = DatasetFinal[all_column_regression]
DatasetRegression.dropna(inplace=True)

s = setup(
    DatasetRegression, # ---> Variabel Dataset
    target='biofuel_consumption', # ---> Target yang mau kita prediksi
    session_id=123, # ---> Randomize Data
    normalize=True,   # ---> Normalize Data (Mengubah scale data menjadi lebih kecil)
    train_size=0.9,   # Training Size = 90%
)
```

```python
# Compare Models OOP API

best = s.compare_models()
```

```python
# OOP API
s.evaluate_model(best) # variabel best -> model terbaik
```

#### 3) Exploratory Data Analysis

```python
plt.figure(figsize=(15,5))
plt.plot(df[df['unique_id']=='FR']['ds'], df[df['unique_id']=='FR']['y'])
plt.xlabel('Date')
plt.ylabel('Price [EUR/MWh]')
plt.grid()
plt.title('Price Time Series')
plt.show()
```

#### 4) Modelling

```python
horizon = 24 # Akan meramalkan data 24 jam kedepan atau 1 hari kedepan
levels = [80,90] # Tingkat keyakinan peramalan
models = [NHITS(h=horizon,
                input_size = 7 * horizon, # Data training adalah 7 hari
                scaler_type = 'robust',
                max_steps = 10,
                loss = MQLoss(level=levels))]
```

```python
nf = NeuralForecast(models=models, freq='H') # H karena hourly, kalau tahunan ganti jadi Y
nf.fit(df=df,
       static_df=static_df)
```

```python
futr_df = pd.read_csv('https://datasets-nixtla.s3.amazonaws.com/EPF_FR_BE_futr.csv')
futr_df['ds'] = pd.to_datetime(futr_df['ds'])
futr_df.head()
```

```python
Y_hat_df = nf.predict(futr_df=futr_df)
Y_hat_df.head()
```

```python
# Visualisasi hasil forecasting

import matplotlib.pyplot as plt
plot_df = df[df['unique_id']=='FR'].tail(24*5).reset_index(drop=True)
Y_hat_df = Y_hat_df.reset_index(drop=False)
Y_hat_df = Y_hat_df[Y_hat_df['unique_id']=='FR']
plot_df = pd.concat([plot_df, Y_hat_df]).set_index('ds') # Concatenate the train and forecast dataframes
plot_df[['y', 'NHITS-median']].plot(linewidth=2)
plt.axvline('2016-11-01', color='red')
plt.ylabel('Price [EUR/MWh]', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.grid()
plt.title('Hasil Forecasting')
```
